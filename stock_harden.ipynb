```python
print("------------开始炼丹----------------")
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('mysql+pymysql://root:210518aA@localhost:3306/stocks')
```

```python
sql_stock = "select * from stocks_dayk_duokong"
df_stock = pd.read_sql_query(sql_stock,engine)
print(df_stock.dtypes)
```

```python
code_gb = df_stock.groupby('code')
```

```python
def biggerlow(df_stock_dayk):
    if df_stock_dayk['low'].shift(1)>df_stock_dayk['close']:
        if df_stock_dayk['low'].shift(2)>df_stock_dayk['close']:
            if df_stock_dayk['low'].shift(3)>df_stock_dayk['close']:
                return '1'
            else :
                return '0'
        else:
            return '0'
    else:
        return '0'
```

```python
for code,code_group in code_gb:
    #code = '300113'
    #print(code)
    df_stock_dayk = code_gb.get_group(code)
    df_stock_dayk = df_stock_dayk.sort_values(by='date',ascending=True)  #单个股票的K线，根据日期排序
    df_stock_dayk['zhangfu'] = df_stock_dayk['close']/df_stock_dayk['close'].shift(1)
    df_stock_dayk['zhangfu_1'] = df_stock_dayk['close'].shift(1)/df_stock_dayk['close'].shift(2)
    df_stock_dayk['changdu'] = df_stock_dayk['close']/df_stock_dayk['open']
    df_stock_dayk['biglow_1'] = df_stock_dayk['low'].shift(-1)>df_stock_dayk['high']
    df_stock_dayk['biglow_2'] = df_stock_dayk['low'].shift(-2)>df_stock_dayk['high']
    df_stock_dayk['biglow_3'] = df_stock_dayk['low'].shift(-3)>df_stock_dayk['high']
    df_stock_dayk['highlow_1'] = df_stock_dayk['high'].shift(-1)<df_stock_dayk['high']*1.12
    df_stock_dayk['highlow_2'] = df_stock_dayk['high'].shift(-2)<df_stock_dayk['high']*1.12
    df_stock_dayk['highlow_3'] = df_stock_dayk['high'].shift(-3)<df_stock_dayk['high']*1.12
    #df_print = df_stock_dayk['date','close','open','zhangfu','zhangfu_1','changdu']
    df_stock_output = df_stock_dayk[(df_stock_dayk['zhangfu']>1.098) & (df_stock_dayk['zhangfu_1']<1.09) & (df_stock_dayk['changdu']>1.05) & (df_stock_dayk['biglow_1'])]
    #df_stock_output = df_stock_output[(df_stock_output['zhangfu']) & (df_stock_output['zhangfu_1']) & (df_stock_output['zhangfu_2'])]
    df_stock_output = df_stock_output[(df_stock_output['biglow_1']) & (df_stock_output['biglow_2']) & (df_stock_output['biglow_3'])]
    df_stock_output = df_stock_output[(df_stock_output['highlow_1']) & (df_stock_output['highlow_2']) & (df_stock_output['highlow_3'])]
    if df_stock_output.shape[0]>0:
        print(code)
        print(df_stock_output['date'])
    #print(df_stock_dayk.tail(10))
    #break
print("----------炼丹完成，祝大吉大利-----------")
```
